# Updating datatype

```python
df['date_reported'] = pd.to_datetime(df['date_reported'], format = '%Y-%m-%d')
df['date_occurred'] = pd.to_datetime(df['date_occurred'], format = '%Y-%m-%d %H:%M:%S') #accounting for time as well
```

```python
df.describe(include = ['float64', 'int64'])
```

```python
df['victim_age'].value_counts()
```

```python
df_age = df[df['victim_age'] > 1]
```

```python
df_age['victim_age'].value_counts()
```

```python
df['Year'] = df['date_occurred'].dt.year
df['Month'] = df['date_occurred'].dt.month
df
```

# Feature understanding

```python
df['Month'].value_counts().sort_index().plot()
```

```python
df['Year'].value_counts().sort_index().plot(kind = 'bar')
```

```python
df = df[~((df['Month'] == 12) & (df['Year'] == 2023))]
df.sort_values('date_occurred')
```

```python
df.groupby(['Year', 'Month']).size().plot()
```

```python
plt.hist(df_age['victim_age'], edgecolor = 'black', bins = 15)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Victim age plot')
```

```python
df['date_occurred_no_time'] = df['date_occurred'].dt.date
df
```

```python
crime_count_per_day = df.groupby(['date_occurred_no_time', 'crime_description']).size()
crime_count_per_day
```

```python
top20_crime = crime_count_per_day.groupby('crime_description').mean()\
                                         .reset_index(name = 'average_daily_count')\
                                        .sort_values(by =  'average_daily_count', ascending = False).head(20)  
top20_crime = top20_crime.sort_values(by =  'average_daily_count')                                                                                                    
                                                                                                             
```

```python
plt.barh(top20_crime['crime_description'], top20_crime['average_daily_count'])
```

```python
plt.figure(figsize = (12,8))

#plot 1
plt.subplot(2,2,1)
crime_plot = df.groupby(['Year', 'Month']).size().plot(color = 'green', linewidth = 2)
crime_plot.set_xlabel('Year-Month', fontsize = 12)
crime_plot.set_title('Crime over Time')


#plot 2 
plt.subplot(2,2,2)
plt.hist(df_age['victim_age'], edgecolor = 'black', bins = 15)
plt.tight_layout()
plt.xlabel('Age', fontsize = 12)

plt.title('Victim age plot')

#plot 3
plt.figure()
plt.barh(top20_crime['crime_description'], top20_crime['average_daily_count'], color = 'skyblue')
plt.xlabel('Frequency', fontsize = 12)
plt.title("Top 20 Crimes")
```

# HOW LONG DOES IT TAKE TO REPORT A CRIME?

```python
time_to_report = df['Time to report'].value_counts().reset_index()
time_to_report.sort_values(by = 'count', ascending = False).head(10)
```

# Which hour has the greatest crime?

```python
hourly_crime = df['date_occurred'].dt.hour.value_counts().sort_index().reset_index()
hourly_crime
```

```python
plt.bar(hourly_crime['date_occurred'], hourly_crime['count'])
plt.xlabel("Hours")
plt.ylabel("Crime")
plt.title("Crime wrt Hours of the day")
```
